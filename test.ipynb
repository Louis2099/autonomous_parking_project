```python
import xml.etree.ElementTree as ET

# Parse the XODR file
tree = ET.parse(r"D:\autonomous_parking_project\maps\xparkinglot_17.xodr")
root = tree.getroot()

# Find all object elements representing parking spaces
parking_spaces = root.findall(".//object[@type='parkingSpace']")

# Extract information for each parking space
for parking_space in parking_spaces:
    id = parking_space.attrib['id']
    s = parking_space.attrib['s']
    t = parking_space.attrib['t']
    hdg = parking_space.attrib['hdg']
    width = parking_space.attrib['width']
    length = parking_space.attrib['length']
    orientation = parking_space.attrib['orientation']
    
    # Process or store the extracted information as needed
    print(f"Parking Space ID: {id}")
    print(f"Position (s, t): ({s}, {t}, {hdg})")
    print(f"Orientation: {orientation}")
    print(f"Width: {width}")
    print(f"Length: {length}")
    print("------------------------")
```
